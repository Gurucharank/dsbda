```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
```

```python
print("Features and Types in the Iris Dataset:")
print(df.dtypes)
```

```python
df.drop(columns='Species').hist(bins=20, figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Each Feature', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 3. Create box plots for each feature
plt.figure(figsize=(10, 8))
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.tight_layout()
plt.show()
```

```python
# 4. Compare distributions and identify outliers
sns.pairplot(df, hue='Species')
plt.suptitle('Pairplot of Iris Dataset', fontsize=16)
plt.tight_layout()
plt.show()
```
